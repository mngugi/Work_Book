```python
import seaborn as sns

sns.set_theme()

fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, kind ="line", x="timepoint", y="signal", col="region",
            hue="event", style="event",)
```

```python
sns.lmplot(data=fmri, x="timepoint", y="signal", col="region",
          )
```

```python
sns.displot(data=fmri, x="timepoint", y="signal", col="region",
          )
```

```python
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
sns.displot(data=tips,kind="ecdf", x="total_bill", col="time", hue="smoker", rug = "True")
```

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker")
```

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
sns.pairplot(data=penguins, hue="species")
```
