```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 0.9
beta = 0.1
gamma = 0.05
iterations = 100
k = 10

# Initialize variables
S = np.zeros(iterations + 1)
T = np.zeros(iterations + 1)
E = np.zeros(iterations + 1)
epsilon = np.random.randn(iterations + 1)

# Initial conditions
S[0] = 1  # Starting state
T[0] = S[0] * k
E[0] = 0  # No initial event

# Simulation loop
for n in range(iterations):
    E[n+1] = np.random.choice([0, 1], p=[0.95, 0.05])  # Random event occurrence
    S[n+1] = alpha * S[n] + beta * E[n] + gamma * epsilon[n]
    T[n+1] = S[n+1] * k

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(S, label='System State (S)')
plt.plot(T, label='Traffic (T)')
plt.plot(E, label='Events (E)')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.legend()
plt.title('Autonomous System Simulation')
plt.show()
```
